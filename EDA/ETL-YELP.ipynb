```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import os
import time

# Ruta del archivo JSON de entrada y del archivo Parquet de salida
input_file = "D:/MaysonPenayo/Escritorio/proyecto final/yelp/review.json"
output_file = "D:/MaysonPenayo/Escritorio/proyecto final/review.parquet"
temp_dir = "D:/MaysonPenayo/Escritorio/proyecto final/temp_parquet"

# Crear el directorio temporal si no existe
os.makedirs(temp_dir, exist_ok=True)

# Contador para archivos temporales
temp_file_counter = 0

# Crear una lista para almacenar los DataFrames temporales
dataframes = []

# Función para guardar DataFrames acumulados en un archivo Parquet temporal
def save_to_temp_parquet(dataframes, temp_dir, temp_file_counter):
    if dataframes:
        table_pa = pa.Table.from_pandas(pd.concat(dataframes, ignore_index=True))
        temp_file = os.path.join(temp_dir, f"temp_{temp_file_counter}.parquet")
        pq.write_table(table_pa, temp_file)
        return temp_file_counter + 1
    return temp_file_counter

# Medir el tiempo de ejecución
start_time = time.time()

# Leer el archivo JSON línea por línea y procesar cada línea
with open(input_file, "r", encoding="utf-8") as file:
    for i, line in enumerate(file):
        try:
            # Leer y convertir cada línea JSON a un DataFrame de pandas
            data = pd.read_json(line, lines=True)
            dataframes.append(data)
        except ValueError as e:
            print(f"Error procesando la línea {i}: {e}")
            continue
        
        # Guardar en Parquet en lotes de 10000 líneas
        if (i + 1) % 10000 == 0:
            temp_file_counter = save_to_temp_parquet(dataframes, temp_dir, temp_file_counter)
            dataframes = []

# Guardar cualquier resto de DataFrames que no alcanzó a formar un lote completo
temp_file_counter = save_to_temp_parquet(dataframes, temp_dir, temp_file_counter)

# Combinar todos los archivos Parquet temporales en uno solo
parquet_files = [os.path.join(temp_dir, f) for f in os.listdir(temp_dir) if f.endswith('.parquet')]
tables = [pq.read_table(f) for f in parquet_files]
final_table = pa.concat_tables(tables)
pq.write_table(final_table, output_file)

# Eliminar archivos temporales
for f in parquet_files:
    os.remove(f)

# Medir el tiempo final y calcular el tiempo transcurrido
end_time = time.time()
elapsed_time = end_time - start_time

print(f"Tiempo total de ejecución: {elapsed_time:.2f} segundos")
print("Proceso completo.")
```

```python
# Especificamos el tamaño del trozo que queremos leer y guardar
chunksize = 1000000  # 1 millón de filas por trozo

# Abre el archivo Parquet
parquet_file = pq.ParquetFile("D:/MaysonPenayo/Escritorio/proyecto final/review.parquet")

# Obtén el número total de grupos de filas en el archivo Parquet
total_row_groups = parquet_file.num_row_groups

# Itera sobre el archivo Parquet y guarda cada trozo como un archivo CSV
for i in range(0, total_row_groups):
    # Lee un trozo del archivo Parquet
    chunk = parquet_file.read_row_group(i, use_pandas_metadata=True).to_pandas()

    # Guarda el trozo como un archivo CSV
    chunk.to_parquet(
        f"D:/MaysonPenayo/Escritorio/proyecto final/chunk_review_{i}.parquet", index=False
    )

    # mensaje de procesos completados por archivo
    print(f"Trozo {i + 1} guardado como chunk_review_{i}.parquet")

print("Proceso completo.")
```
