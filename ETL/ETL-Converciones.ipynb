```python
import pandas as pd
```

```python
# Leer el archivo .parquet y cargarlo en un DataFrame de pandas
df = pd.read_parquet('Business_Category.parquet')
# Guardar el DataFrame como un archivo .csv sin incluir el índice
df.to_csv('Business_Category.csv', index=False)
```
