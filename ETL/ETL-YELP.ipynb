Las categorias que son nulas, las cambiamos por una cadena vacia y creamos una lista con cada categoria unica

```python
dfBusiness['categories'] = dfBusiness['categories'].fillna('')
# Dividir las cadenas de texto en listas de categorías
categories_list = dfBusiness['categories'].str.split(', ')

# Aplanar la lista de listas a una sola lista
all_categories = [item for sublist in categories_list for item in sublist]

# Obtener las categorías únicas
unique_categories = list(set(all_categories))
```

```python
categories_df = pd.DataFrame({
    'id': range(1, len(unique_categories) + 1),
    'category': unique_categories
})
```

```python
categories_df
```

```python
categories_df[categories_df["category"] == "Shopping"]
```

Iteraramos sobre cada negocio y sus categorías para crear un df donde se tiene cada negocio con sus respectivas categorias

```python
# Crear un diccionario para mapear categorías a category_id
category_id_map = dict(zip(categories_df['category'], categories_df['id']))

# Crear una lista para almacenar los datos finales
business_category_pairs = []

# Iterar sobre cada negocio y sus categorías
for idx, row in dfBusiness.iterrows():
    business_id = row['business_id']
    categories = row['categories'].split(', ')
    for category in categories:
        if category:  # Ignorar cadenas vacías
            category_id = category_id_map[category]
            business_category_pairs.append({'business_id': business_id, 'category_id': category_id})

# Crear el DataFrame final
business_category_df = pd.DataFrame(business_category_pairs)
```
